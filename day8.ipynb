```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

#### <u> Detecting Outliers
<ol>
    <li>Only for the numerical column</li>
    <li>Identified via visualization - Using Box plot</li>
</ol>

```python
# chart types- univariate , multivariate
movie_data = pd.read_csv('movies.csv')
```

```python
#steps to make chart for outlier
# step 1 -> set the size of figure
plt.figure(figsize=(10,6))


# step 2 -> create the chart
sns.boxplot(x='RATING' , y='RunTime' , data=movie_data )

# step 3 -> label the Chart
plt.xlabel('Rating')
plt.ylabel('Runtime')
plt.title('Boxplot- Runtime by Rating')
plt.xticks(rotation=45)

# step 4-> confirm the configuration and print chart
plt.show()
```

<h1>Creating hitmap for corelation</h1>
 <ol>
     <li>only for numerical colums</li>
     <li>Visualitaion for Corelation matrix</li>
 </ol>

```python
# Drop rows with NaNs (optional, depends on your data)
movie_data.dropna(subset=['RATING', 'RunTime', 'VOTES'], inplace=True)
# Clean numeric columns
movie_data['VOTES'] = movie_data['VOTES'].str.replace(',', '').astype(float)
movie_data['RunTime'] = pd.to_numeric(movie_data['RunTime'], errors='coerce')
movie_data['RATING'] = pd.to_numeric(movie_data['RATING'], errors='coerce')
```

```python
plt.figure(figsize=(10,6))
corr_mat=movie_data[['RATING','RunTime','VOTES']].corr()
sns.heatmap(corr_mat , annot= True , cmap='coolwarm')
plt.title("Corelation Matrix")

print(corr_mat)
```
